```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
import scipy.cluster.hierarchy as shc
from sklearn.cluster import AgglomerativeClustering
%matplotlib inline
```

```python
data = pd.read_csv('data.csv')
data.head()
```

There are multiple product categories â€“ Fresh, Milk, Grocery, etc. The values represent the number of units purchased by each client for each product.

Using Clustering we can group similar clients together.

```python
normalized_data = pd.DataFrame(normalize(data), columns = data.columns)
normalized_data.head()
```

```python
plt.figure(figsize=(10, 7))  
plt.title('Dendrogram by Single Linkage')
plt.xlabel('Samples')
plt.ylabel('Euclidean distances')
dendogram_by_single = shc.dendrogram(shc.linkage(normalized_data, method='single'))
```

```python
plt.figure(figsize=(10, 7))  
plt.title('Dendrogram by Complete Linkage')
plt.xlabel('Samples')
plt.ylabel('Euclidean distances')
dendogram_by_complete = shc.dendrogram(shc.linkage(normalized_data, method='complete'))
```

```python
plt.figure(figsize=(10, 7))  
plt.title('Dendrogram by Ward Linkage')
plt.xlabel('Samples')
plt.ylabel('Euclidean distances')
dendogram_by_ward = shc.dendrogram(shc.linkage(normalized_data, method='ward'))
```

```python
cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean', linkage='ward')  
cluster.fit_predict(normalized_data)
```

```python
print(cluster.labels_)
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(normalized_data['Milk'], normalized_data['Grocery'], c=cluster.labels_)
```

```python
from sklearn.cluster import Birch
b_cluster = Birch(n_clusters=2, threshold=0.025)
```

```python
b_cluster.fit_predict(normalized_data)
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(normalized_data['Milk'], normalized_data['Grocery'], c=b_cluster.labels_)
```
