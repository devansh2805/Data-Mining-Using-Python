```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
iris = datasets.load_iris()
X = pd.DataFrame(iris.data, columns=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width'])
y = pd.DataFrame(iris.target, columns=['Target'])
```

```python
sqd_error = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(X)
    sqd_error.append(kmeans.inertia_)
plt.plot(range(1, 11), sqd_error)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Squared Error')
plt.show()
```

```python
iris_k_mean_model = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
y_kmeans = iris_k_mean_model.fit_predict(X)
x = X.values
```

```python
plt.figure(figsize=(12,3))
plt.scatter(x[y_kmeans == 0, 2], x[y_kmeans == 0, 3], c = 'red', label='setosa')
plt.scatter(x[y_kmeans == 1, 2], x[y_kmeans == 1, 3], c = 'blue', label='versicolor')
plt.scatter(x[y_kmeans == 2, 2], x[y_kmeans == 2, 3], c = 'green', label='virginica')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.legend()
plt.show()
```

```python
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], c='red', label='setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], c='blue', label='versicolor' )
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], c='green', label='virginica')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.legend()
plt.show()
```
