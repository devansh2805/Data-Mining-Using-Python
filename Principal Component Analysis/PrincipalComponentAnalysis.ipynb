```python
import pandas as pd
import numpy as np
```

```python
covid19_data = pd.read_csv('covid19.csv', na_values = [''])
covid19_data.dropna(axis = 0, inplace = True)
X = covid19_data.iloc[:, 2:9]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 4)
X = pca.fit_transform(X)
print(X)
```
