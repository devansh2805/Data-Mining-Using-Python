```python
# Importing Libraries
import pandas as pd
import numpy as np
import seaborn as sns
```

Loading Dataset, Tips.csv is a dataset comparing tips given based on total bill and other attributes such as day of the week, Time of the day, whether the customer is smoker or not and size of group

```python
# Loading Dataset
tips_data = pd.read_csv('Tips.csv', index_col = 0, na_values = ['NaN'])
tips_data.dropna(axis = 0, inplace = True)
print(tips_data)
sns.set(style = "darkgrid")
```

```python
# Regression Plot
sns.regplot(x=tips_data['TotalBill'], y=tips_data['Tips'])
```

Regression Plot for TotalBill (Independent Variable) v/s Tips (Dependent Variable). Visualization shows Positive co-relation between both, i.e. if total bill increases tip also increases

```python
# Scatter Plot
sns.regplot(x=tips_data['TotalBill'], y=tips_data['Tips'], fit_reg=False)
```

Scatter Plot for Total Bill and Tips

```python
# Scatter Plt with Category
sns.lmplot(x='TotalBill', y='Tips', data=tips_data, hue='Time', fit_reg=False, legend=True)
```

Scatter Plot of TotalBill v/s Tips with Time as Category, we can see most tips are given during Dinner time, also Higher tips are also given at Dinner

```python
# Histogram with Gaussian Distribution
sns.histplot(tips_data['Tips'], kde=True)
```

Histogram Plot (Frequency Plot) of Tips. We can see maximum Tips are from 2 to 4

```python
# Histogram without Gaussian Distribution
sns.histplot(tips_data['Tips'], kde=False)
```

Frequency Plot without Gaussian Distribution

```python
# Histogram with Fixed Number of Bins
sns.histplot(tips_data['Tips'], kde=True, bins=5)
```

Histogram Plot for Tips with Fixed Bins

```python
# Bar Plot
sns.countplot(x='Day', data=tips_data)
```

Bar Graph showing Data for Number of tips given at each day, we can see maximum tips were given on saturday and least on friday

```python
# Bar Plot with Category
sns.countplot(x='Day', data=tips_data, hue='Smoker')
```

Bar Graph showing Data for Number of tips given at each day with also categorising each day with customer's smoking attribute, we can see more tips are given by non smokers

```python
# Box Plot
sns.boxplot(y=tips_data['Tips'])
```

Box Plot Analysis for Tips, we can see there are'nt many outliers in the data

```python
# Multiple Box Plot
sns.boxplot(x=tips_data['Size'],y=tips_data['Tips'])
```

Box Plot analysis for tips with Group Size, we can see group size 4 has higher tips than group size 1

```python
# Grouped Box Plot
sns.boxplot(x=tips_data['Size'],y=tips_data['Tips'], hue=tips_data['Smoker'])
```

Box plot analysis for tips with group size categorized with customer's smoking attribute, again we can safely conclude higher groupsize non smokers give higher tips
